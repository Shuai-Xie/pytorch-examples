```python
# %matplotlib inline
%matplotlib widget
import tensorwatch as tw
client = tw.WatcherClient(filename='test.log')
```

```python
s0 = client.open_stream(name='metric1')
v0 = tw.Visualizer(stream=s0)
v0.show()
```

```python
import torchvision

alexnet_model = torchvision.models.alexnet()
tw.draw_model(alexnet_model, input_shape=[1,3,224,224])
```

```python
# 非常方便查看 模型每层参数，以及执行时间
# 最后1行(21) 给出了完整的统计信息
tw.model_stats(alexnet_model, input_shape=[1,3,224,224])
```
